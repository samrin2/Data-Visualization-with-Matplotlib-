```python
# Bar Chart: Total number of households per year
plt.figure(figsize=(10, 6))
plt.bar(df['year'], df['tot_hhs'], color='skyblue')
plt.xlabel('Year')
plt.ylabel('Total Households')
plt.title('Total Number of Households per Year')
plt.xticks(df['year'], rotation=45)
plt.show()
```

```python
# Line Chart: Proportion of households owning washing machines over the years
plt.figure(figsize=(10, 6))
plt.plot(df['year'], df['own_wm_prop'], marker='o', linestyle='-', color='green')
plt.xlabel('Year')
plt.ylabel('Proportion of Households Owning Washing Machines (%)')
plt.title('Proportion of Households Owning Washing Machines Over the Years')
plt.grid(True)
plt.show()
```
